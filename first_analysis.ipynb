```r
mydata2 = mydata %>%
 rename(bads_sf = 'BADS-SF',
       activation = 'BADS-SF_Activation',
       avoidance = 'BADS-SF_Avoidance',
       ces_d = 'CES-D',
       whoqol_mean = 'WHOQOL-26_Mean total score',
       whoqol_physical = 'WHOQOL-26_Phisical health',
       whoqol_psych = 'WHOQOL-26_Psychological health',
       whoqol_soc = 'WHOQOL-26_Social relationships',
       whoqol_env = 'WHOQOL-26_Environment',
       whoqol_overall = 'WHOQOL-26_Overall QOL')

head(mydata2)
```

```r
# Working with grammars: select
# Let's say we want to work with only BADS SF, BADS Activation, BADS Avoidance, and CES Depression scores 
# and all Gender and Age
mydata_short = mydata2 %>%
                select(Sex:ces_d)
head(mydata_short)
```

```r
## Grammar: filter
teens_f = mydata2 %>%
   filter(Sex == 2, 
         Age < 20)
head(teens_f)
```

```r
# Grammar: arrange
young_first = mydata2 %>%
               arrange(Age)
head(young_first) # youngest person is aged 2!
```

```r
# Grammar: mutate
# create a new variable out of Age by cutting it to 4 equally spaced age groups
# More information: https://ggplot2.tidyverse.org/reference/cut_interval.html
mydata3 = mydata2 %>%
           mutate(age_rec = cut_interval(Age, n = 4))
# head(mydata3)
# check it has done so, let's count the groups:
mydata3 %>%
 count(age_rec)

typeof(mydata2$Age)
```

```r
mydata4 = mydata2 %>%
  mutate(age_bin = cut_number(Age, n = 2,
                             label = c("2-19", "> 20")))


mydata4 %>%
 count(age_bin)
            
```

```r
# What is the average age of the respondents?
average_age = mydata4 %>%
 summarise(average_age = mean(Age, na.rm = T))
average_age
```

```r
# Are males older than females?
av_age = mydata4 %>%
     group_by(Sex) %>%
     summarize(average_age = mean(Age, na.rm = T))
```

```r
av_age
```

```r
# Do teenagers have lower average QoL than older people?
qol_score = mydata4 %>%
     group_by(age_bin) %>%
     summarize(qol_score = mean(whoqol_mean, na.rm = T))
     
```

```r
qol_score
```

```r
age_sex = mydata4 %>%
     count(Sex, age_bin)
     
age_sex     
```

```r
age_sex_spread = age_sex %>%
 spread(key = Sex, value = n)

age_sex_spread
```

```r
age_sex_df = age_sex_spread %>%
                 rowwise() %>%
                mutate(male_pct = `1` * 100 / sum(`1` + `2`))
age_sex_df                
```

```r
age_sex2 = mydata4 %>%
  group_by(Sex, age_bin) %>%
  summarize(n_age_bin = n())
```

```r
age_sex2
```

```r
## Let's create a small data set for 
small_set = mydata4 %>%
  select(age_bin, Sex, whoqol_mean:whoqol_env)
```

```r
head(small_set) # take a look at the head
```

```r
# Let's group and summarize the whoqol scores
whoqol_scores = small_set %>%
           group_by(age_bin, Sex) %>%
           summarize(qol = mean(whoqol_mean, na.rm = T),
                     physical = mean(whoqol_mean, na.rm = T),
                     psych = mean(whoqol_psych, na.rm = T),
                     soc = mean(whoqol_soc, na.rm = T),
                     env = mean(whoqol_env, na.rm = T)
                     )
```

```r
whoqol_scores
```

```r
whoqol_scores_gathered = whoqol_scores %>%
        gather(key = qaulity_of_life, qol, physical, psych, soc, env, value = all_means)
```

```r
whoqol_scores_gathered
```

```r
whoqol_sex = whoqol_scores_gathered %>%
  spread(key = Sex, value = all_means)
```

```r
whoqol_sex
```

```r
whoqol_age = whoqol_scores_gathered %>%
  spread(key = age_bin, value = Averages)
```

```r
mydata4 %>%
  ggplot() +
  geom_bar(aes(x = Age)) +
  facet_wrap(~Sex) +
  ggtitle("A bar plot of age by gender") +
  xlab("Age in years") +
  ylab("Counts") +
  ggsave("age_bar_sex.png")
```

```r
mydata4 %>%
  mutate(gender = as.factor(Sex)) %>%
  ggplot() +
  geom_point(aes(x = bads_sf, y = whoqol_mean, colour = gender)) +
  geom_smooth(aes(x = bads_sf, y = whoqol_mean, colour = gender), method = "lm") +
  ggtitle("Association between bads_sf and whoqol") +
  xlab("BADS-SF score") +
  ylab("WHOQoL Score") +
  ggsave("bads_who_lm.png")
```
